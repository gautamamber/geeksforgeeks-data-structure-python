```python
def toh(n, A, B, C):
    if n == 1:
        print("Move 1 from", A, " to ", C)
    else:
        toh(n-1, A, C, B)
        print("Move", n, "from", A , " to ", C)
        toh(n-1, B, A, C)
```

```python
toh(2, 'A', 'B', 'C')
```

```python
toh(3, 'A', 'B', 'C')
```

```python
toh(4, 'A', 'B', 'C')
```
